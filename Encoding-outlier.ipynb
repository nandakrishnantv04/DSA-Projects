## Outliers

```python
df= pd.read_csv(r"C:\Users\Lenovo\OneDrive\Desktop\nandu\shokam\ictak\House_Pricing.csv")
```

```python
for i in ['Sale Price','No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)','Area of the House from Basement (in Sqft)','Zipcode', 'Latitude',
       'Longitude', 'Living Area after Renovation (in Sqft)']:
    df[i] = df[i].fillna(df[i].median())
```

```python
plt.boxplot(df['No of Bathrooms'])
plt.show
```

```python
Q1 = np.percentile(df['No of Bathrooms'],25)
```

```python
Q1
```

```python
Q3 = np.percentile(df['No of Bathrooms'],75)
```

```python
Q3
```

```python
IQR = Q3-Q1
```

```python
IQR
```

```python
low_lim = Q1 - 1.5*IQR
upr_lim = Q3 + 1.5*IQR
```

```python
low_lim
```

```python
upr_lim
```

```python
outlier = []
for x in df['No of Bathrooms']:
    if x>upr_lim or x<low_lim:
        outlier.append(x)
```

```python
outlier
```

```python
ind = df['No of Bathrooms']>upr_lim
df.loc[ind].index
```

```python
df = df[(df['No of Bathrooms'] >= low_lim) & (df['No of Bathrooms'] <= upr_lim)]
```
