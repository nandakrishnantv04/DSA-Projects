```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
###Visualization 1: Histogram of Petal Length (PL)
plt.figure(figsize=(8, 6))
sns.histplot(df['PL'], bins=20, color='green')
plt.title('Distribution of Petal Length (PL)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
###Visualization 2: Box Plot of Petal Length by Classification
plt.figure(figsize=(10, 7))
sns.boxplot(x='Classification', y='PL', data=df)
plt.title('Petal Length Distribution by Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
###Visualization 3: Scatter Plot of Petal Length vs. Petal Width, Colored by Species
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PL', y='PW', hue='Classification', data=df, style='Classification', s=100)
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```
